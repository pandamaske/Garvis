```python
from diffusers import StableDiffusionPipeline, DDIMScheduler
import torch

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Load the Stable Diffusion model
pipe = StableDiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-2-base")
pipe.to(device)
# Example Stable Diffusion prompt in Python

prompt = (
    "A modern, minimalistic logo for an HCM platform called ‘Hunite’ (Humans United Network Information Systems). "
    "The logo should feature the letters H, U, and I. Design the ‘H’ to represent two people shaking hands, symbolizing partnership and collaboration. "
    "The ‘U’ should form a smile connecting the bottoms of the ‘H’ bars, representing satisfaction and positivity. "
    "The ‘I’ should be placed subtly behind the ‘H’, symbolizing support and integration. The design should be sleek, simple, and professional, emphasizing unity and trust. "
    "Use the colors #ffffff (white), #3200e2 (blue), and #eb692b (orange) to create a clean and impactful design."
)

negative_prompt = (
    "complex designs, overly intricate details, clutter, "
    "using gradients, shadows, or any 3D effects."
)


# Example inference settings
num_inference_steps = 50
guidance_scale = 7.5  # often called 'CFG Scale' in some UIs


image = pipe(
   prompt=prompt,
     negative_prompt=negative_prompt,
     guidance_scale=guidance_scale,
    num_inference_steps=num_inference_steps
 ).images[0]

#Save or display your image
image.save("output.png")
#
print("Image generated and saved!")





```

```python
from diffusers import StableDiffusionPipeline, DDIMScheduler
import torch

!huggingface-cli login

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Load the Stable Diffusion model
pipe = StableDiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-2-base")
pipe.to(device)
# Define a detailed prompt
prompt = (
    "a Jewish shinigami in the Style of the manga bleach by tite kubo"
)
# Configure the scheduler
scheduler = DDIMScheduler(
    beta_start=0.00085,
    beta_end=0.012,
    beta_schedule="scaled_linear"
)
# Generate the image
image = pipe(
    prompt,
    scheduler=scheduler,
    num_inference_steps=50,
    guidance_scale=7.5,
    torch_dtype=torch.float16
).images[0]
# Save the generated image
image.save("paris_ruins.png")
```

```python
import torch
from diffusers import FluxPipeline
!huggingface-cli login

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Load the Stable Diffusion model

pipe = FluxPipeline.from_pretrained("black-forest-labs/FLUX.1-dev", torch_dtype=torch.bfloat16)
pipe.to(device)

prompt = "A cat holding a sign that says hello world"
image = pipe(
    prompt,
    height=1024,
    width=1024,
    guidance_scale=3.5,
    num_inference_steps=50,
    max_sequence_length=512,
    generator=torch.Generator("cpu").manual_seed(0)
).images[0]
image.save("flux-dev.png")
```
